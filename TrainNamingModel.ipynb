# Upload Training file

OK, ready to start. Run the first code section below. 

You will be prompted to upload a training file.

This file **MUST** be a plain text file with one word per line. The name is not important as it will be renamed after uploading as long as the extension is 'txt'

```
import os
from google.colab import files

WORDLIST = "words.txt"
MODEL = "model.hdf5"

import os
if os.path.exists(WORDLIST):
  os.remove(WORDLIST)
  
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

if len(uploaded)!=1:
  print("üö® Error, you must only upload one word list file.")
else:
  for fn in uploaded.keys():
    if os.path.splitext(fn)[1] == ".txt":
       os.rename(fn,WORDLIST)
       print("Renamed {0} to {1}".format(fn,WORDLIST))
```

# Set Parameters for the Neural Network

Select the parameters for the network. **Recommended to keep the defaults**


1.   **layers** - defaults to 2, more layers will give a better but slower network
2.   **size** - defaults to 128, how many nodes in the network, more nodes gives a better but slower network
3.   **epochs** - how many iterations of the network to run, more epochs gives better results but a slower network



```
#@title Set the number of layers, size and number of epochs for the network { run: "auto" }

layers = 2 #@param ["1", "2", "3"] {type:"raw"}
size =  128 #@param ["64", "128", "256"] {type:"raw"}
epochs = 10 #@param ["5", "10", "15"] {type:"raw"}
print('You selected {} layers'.format(layers))
print('You selected a size of {}'.format(size))
print('You selected {} epochs'.format(epochs))
```

# Run the Model Generator

This will generate the model and save it as a file to be downloaded in the next step. Grab a ‚òï as this may take a little while!

```
from textgenrnn import textgenrnn

t = textgenrnn(name=MODEL)
t.train_from_file(WORDLIST,rnn_layers=layers,rnn_size=size, num_epochs=epochs)
print(t.model.summary())

print("‚úîÔ∏è Finished generating {}.".format(MODEL))
```

# Download the Generated Model

Running the code below will use your browser to download a file containing the generated model to your computer.

```
if os.path.exists(MODEL):
  files.download(MODEL)
  print("‚úîÔ∏è Finished downloading file {}".format(MODEL))
else:
  print("üö® Model file {} cannot be found, make sure you run all code before this point first".format(MODEL))
```
